```python
import plotly.graph_objects as go

# Compute averages by gender & job
avg_salary_gender = df.groupby(['Job Title', 'Gender'])['Salary'].mean().reset_index()

# Pivot for dumbbell structure
pivot = avg_salary_gender.pivot(index='Job Title', columns='Gender', values='Salary').dropna()
pivot = pivot.reset_index()

fig4 = go.Figure()

fig4.add_trace(go.Scatter(
    x=pivot['Male'],
    y=pivot['Job Title'],
    mode='markers',
    name='Male',
    marker=dict(color='blue', size=10)
))

fig4.add_trace(go.Scatter(
    x=pivot['Female'],
    y=pivot['Job Title'],
    mode='markers',
    name='Female',
    marker=dict(color='red', size=10)
))

for i, row in pivot.iterrows():
    fig4.add_trace(go.Scatter(
        x=[row['Male'], row['Female']],
        y=[row['Job Title'], row['Job Title']],
        mode='lines',
        line=dict(color='grey', width=2),
        showlegend=False
    ))

fig4.update_layout(
    title='Dumbbell Plot: Avg Salary by Gender & Job Title',
    xaxis_title='Average Salary',
    yaxis_title='Job Title',
    height=800
)

fig4.show()
```

```python
import plotly.graph_objects as go

# Create salary range bins
salary_max = df['Salary'].max()
if salary_max <= 250000:
    salary_max += 1  # Ensure unique bin

df['Salary Range'] = pd.cut(
    df['Salary'],
    bins=[0, 50000, 100000, 150000, 200000, 250000, salary_max],
    labels=['<50k', '50k-100k', '100k-150k', '150k-200k', '200k-250k', f'>{int(salary_max)}']
)

# Group for counts
source_labels = []
target_labels = []
values = []

# Ed Level  Job Title
ed_job = df.groupby(['Education Level', 'Job Title']).size().reset_index(name='Count')

source_labels.extend(ed_job['Education Level'])
target_labels.extend(ed_job['Job Title'])
values.extend(ed_job['Count'])

# Job Title  Salary Range
job_sal = df.groupby(['Job Title', 'Salary Range']).size().reset_index(name='Count')

source_labels.extend(job_sal['Job Title'])
target_labels.extend(job_sal['Salary Range'].astype(str))
values.extend(job_sal['Count'])

# Make unique labels
labels = list(pd.unique(source_labels + target_labels))
source = [labels.index(s) for s in source_labels]
target = [labels.index(t) for t in target_labels]

fig5 = go.Figure(go.Sankey(
    node=dict(label=labels),
    link=dict(source=source, target=target, value=values)
))

fig5.update_layout(title_text='Sankey Diagram: Education  Job  Salary Range', font_size=10)
fig5.show()
```

##  **Interpretation**
* Career pathways: You can see if Bachelors graduates cluster in mid-level roles, or if Masters funnel more into high-paying ones.
* Salary bands: Visualize which education/job combinations tend to stay in lower or higher salary ranges.
* Insight: Great for workforce planning  e.g., identify where more advanced training or promotions might be needed.

```python
# ------------------------------------
#  1 Standardize Education Levels
# ------------------------------------

# Define common replacements
education_map = {
    "Bachelor's": "Bachelor's Degree",
    "Bachelors": "Bachelor's Degree",
    "Master's": "Master's Degree",
    "Masters": "Master's Degree",
    "PhD": "Doctorate",
    "phd": "Doctorate",
    "High School": "High School Diploma"
}

# Apply the map
df['Education Level'] = df['Education Level'].replace(education_map)

# ------------------------------------
#  2 Clean and standardize Job Titles
# ------------------------------------

def clean_job_title(title):
    if pd.isnull(title):
        return title  # Keep NaN if present

    t = title.lower()

    # Expand common abbreviations
    t = t.replace('sr.', 'senior').replace('sr', 'senior')
    t = t.replace('mgr', 'manager')
    t = t.replace('vp', 'vice president')
    t = t.replace('svp', 'senior vice president')
    t = t.replace('avp', 'assistant vice president')

    # Remove dots and extra spaces
    t = t.replace('.', '').strip()

    # Title case for final version
    t = t.title()

    return t

df['Job Title Clean'] = df['Job Title'].apply(clean_job_title)

# ------------------------------------
#  3 Quick check
# ------------------------------------

print(df[['Job Title', 'Job Title Clean']].drop_duplicates().head(15))
print(df['Education Level'].unique())

# ------------------------------------
#  4 Save cleaned version if needed
# ------------------------------------

df.to_csv('Salary_Data_Cleaned.csv', index=False)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.neighbors import KNeighborsRegressor
from xgboost import XGBRegressor

models = {
    "LinearRegression": LinearRegression(),
    "Ridge": Ridge(alpha=1.0),
    "RandomForest": RandomForestRegressor(n_estimators=100, random_state=42),
    "KNN": KNeighborsRegressor(),
    "GradientBoosting": GradientBoostingRegressor(n_estimators=100, random_state=42),
    "XGBoost": XGBRegressor(n_estimators=100, random_state=42)
}

results = {}

for name, model in models.items():
    pipe = Pipeline([
        ('model', model) 
    ])
    
    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)
    
    rmse = mean_squared_error(y_test, y_pred, squared=False)
    r2 = r2_score(y_test, y_pred)
    
    results[name] = {'RMSE': rmse, 'R': r2}
    
    print(f"{name}")
    print(f"  RMSE: {rmse:.2f}")
    print(f"  R: {r2:.4f}")
    print("-" * 50)


best = min(results.items(), key=lambda x: x[1]['RMSE'])
print(f"=> Best Model: {best[0]} with RMSE: {best[1]['RMSE']:.2f} and R: {best[1]['R']:.4f}")
```

```python
from sklearn.model_selection import GridSearchCV
import joblib
param_grid = {
    'model__n_estimators': [100, 200, 300],
    'model__max_depth': [None, 5, 10],
    'model__min_samples_split': [2, 5, 10],
    'model__min_samples_leaf': [1, 2, 4]
}
pipeline = Pipeline([
    ('model', RandomForestRegressor(random_state=42))
])
```

```python
grid = GridSearchCV(
    estimator=pipeline,
    param_grid=param_grid,
    cv=5,
    n_jobs=-1,
    scoring='neg_root_mean_squared_error'
)

grid.fit(X_train, y_train)

print(f"Best params: {grid.best_params_}")
print(f"Best CV RMSE: {abs(grid.best_score_):.2f}")
```

```python
best_model = grid.best_estimator_

y_pred = best_model.predict(X_test)

rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)

print(f"Test RMSE: {rmse:.2f}")
print(f"Test R: {r2:.4f}")
```
